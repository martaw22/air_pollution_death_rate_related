```python
data = pd.read_csv("data_air/daily_aqi_by_county_2017.csv")
```

# A. Data Preparation  

```python
data["state_county"] = data["county Name"] + "_" +  data["State Name"]
data.head()
```

```python
## test for one state 
df_state = data[data["state_county"] == "Baldwin_Alabama"]
df_state["which_date"] = df_state["Date"].apply(lambda x: pd.to_datetime(x))
df_state["current_month"] = df_state["which_date"].apply(lambda x: x.month)
df_state["current_date"] = df_state["which_date"].dt.day
df_state['day_of_week'] = df_state['which_date'].dt.weekday_name
day_df = pd.get_dummies(df_state["day_of_week"], prefix="day")
df_temp = pd.concat([df_state, day_df], axis = 1)
df_feature = df_temp[list(day_df.columns) + ["AQI", "current_date", "current_month", "which_date"]]
df_feature = df_feature.sort_values(by=["which_date"])
df_feature
```

```python
# make sure that there is only 1 AQI for each day 
df_state.groupby("Date").count().unstack().reset_index().rename(columns={0: "count"})["count"].max()
```

```python
lag_time = 30;
temps = df_feature.sort_values(by=["which_date"])["AQI"]
dataframe = temps
col_name=['AQI']
for lag_index in range(1,lag_time):
    dataframe = concat([temps.shift(lag_index), dataframe], axis=1)
    col_name.append('lag_' + str(lag_index))
dataframe.columns = col_name
df_lag_features = dataframe.iloc[lag_time-1:,:]
df_lag_features.head()
```

```python
df_data = (pd.concat([df_lag_features.drop(["AQI"], axis = 1), 
                     df_feature.drop(["which_date"], axis = 1).iloc[lag_time-1:,:]], axis = 1))
print(df_data.shape)
df_data.head()
```

```python
## export csv file 
df_data.to_csv("data_air/Baldwin_Alabama_features_2018.csv", index = False)
df_data.head()
```
